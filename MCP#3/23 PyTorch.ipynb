# 🟦Autograd(자동미분)

- `torch.autograd` 패키지는 Tensor의 모든 연산에 대해 **자동 미분** 제공
  - '기울기' 를 주어 학습이 이루어지게 함
- 이는 코드를 어떻게 작성하여 실행하느냐에 따라 역전파(back propagation)가 정의된다는 뜻
- `backprop`를 위해 미분값을 자동으로 계산

## requires_grad= 속성
`requires_grad` 속성을 `True`로 설정하면, 해당 텐서에서 이루어지는 모든 연산들을 추적하기 시작

기록을 추적하는 것을 중단하게 하려면, `.detach()`를 호출하여 연산기록으로부터 분리

```python
torch.manual_seed(42)  # random seed
x = torch.randn(3, 3)
print(x)
print(x.requires_grad)  # requires_grad 값은 기본적으로 False

x = x * 3  # 텐서에 x 3 한 값
print(x)
```

```python
"""
requires_grad 값을 True 로 바꾸면 ?
"""
None
```

`requires_grad_(...)`는 기존 텐서의 `requires_grad` 값을 바꿔치기(`in-place`)하여 변경

`grad_fn`: 미분값을 계산한 함수에 대한 정보 저장 (어떤 함수에 대해서 backprop 했는지)

```python
x.requires_grad = True
```

```python
x.requires_grad
```

```python
x.grad_fn   # 미분값을 계산한 함수에 대한 정보. x 의 grad_fn은 None 이다.
```

```python
# 다음과 같은 연산을 x 에 대해 수행해본다
y = (x * x).sum()
y
```

```python
y.grad_fn # grad_fn= sum 연산을 기록하고 있다!
```

```python
# 어떤 연산에 대해 back propagation 을 했는지 기록하는게 grad_fn
```

## 기울기(Gradient)

```python
# requires_grad=True 로 tensor 생성.
x = torch.ones(3, 3, requires_grad=True)
x
```

```python
x.grad_fn
```

```python
# 이제 x 의 연산은 gradient 추적이 된다

y = x + 5  # x 에 덧셈 연산한 결과 y
print(y)  # x로부터 덧셈한 결과라는 '기록' 이 쌓임.
```

```python
# 위 y 값에 또 연산을 수행해보자
z = y * y   # 연산
out = z.mean() # 또 연산

print(z)
print(out)
```

### .backward()  와 .grad속성
계산이 완료된 후, `.backward()`를 호출하면 자동으로 역전파 계산이 가능하고, `.grad` 속성에 누적됨

```python
x.grad  # .grad : 미분값, 기울기

# 지금은 계산된게 없다.
# 잠시후 backward() 후에 계산된 것을 보게 될거다!
```

```python
# 너무나도 편하게 알아서 미분을 수행해준다!
print(out)
out.backward()  # 역전파 계산!
```

### .grad
`grad`: data가 거쳐온 layer에 대한 미분값 저장

```python
print(x)
print(x.grad)  # 위에서 backward() 한 결과로 x 의 미분값이 자동 계산되어 있다!
```

### with torch.no_grad()

`with torch.no_grad()`를 사용하여 기울기의 업데이트를 하지 않음

기록을 추적하는 것을 방지하기 위해 코드 블럭을 `with torch.no_grad()`로 감싸면 기울기 계산은 필요없지만, `requires_grad=True`로 설정되어 학습 가능한 매개변수를 갖는 모델을 **평가(evaluate)**할 때 유용

```python
# Tensorflow : Define and Run 방식
# PyTorch : Define By Run 방식

# PyTorch 에선 train mode 와 test mode 가 있다.
# '학습모드(train mode)' 에서는 기울기 구하는게 가능하지만
# '테스트모드(test mode)' 에서는 이기능을 꺼야 겠죠
#    -> no_grad() 사용

# no_grad() 이면  기울기를 구하지 않게 됩니다.
# 따라서 batch normalization, dropout 들이 작동 안함
# 작동 속도는 train mode 일때보다 test mode 가 더 빠르겠죠.
```

```python
x.requires_grad
```

```python
(x ** 2).requires_grad  # True
```

```python
with torch.no_grad():  # 기울기 계산 안함!
  print((x ** 2).requires_grad) # 여기선 False 다!
```

### .detach()
`detach()`: 내용물(content)은 같지만 `require_grad`가 다른 새로운 Tensor를 가져올 때

```python
print(x.requires_grad)  # True

# x.numpy()

w = x.detach()  # x에서 detach 된 w 는 더이상 gradient 계산을 안한다
print(w.requires_grad)  # False

print(x)
print(w)
print(w.numpy())
```

## transform 정의
- `DataLoader`의 인자로 들어갈 `transform`을 미리 정의할 수 있고, `Compose`를 통해 리스트 안에 순서대로 전처리 진행

- `ToTensor`()를 하는 이유는 `torchvision`이 PIL Image 형태로만 입력을 받기 때문에 데이터 처리를 위해서 Tensor형으로 변환 필요

  1. 자료형 변환

    - 입력: PIL.Image 또는 NumPy ndarray (dtype=uint8, 0~255)

    - 출력: torch.FloatTensor (dtype=float32)

  1. 차원 순서 변경

    - 일반 이미지: (높이, 너비, 채널) → PyTorch 텐서: (채널, 높이, 너비)

    - 예: (H=224, W=224, C=3) → (C=3, H=224, W=224)

  1. 값 범위 스케일링

    - 0 ~ 255 정수 값 → 0 ~ 1 실수 값

    - 즉, pixel_value / 255.0

```python
# transform 정의
mnist_transform = transforms.Compose([
    transforms.Resize((28, 28)),  # 사실 원본이 이미 28, 28
    transforms.ToTensor(), # 타입변환, 차원순서변경, 스케일링 -> [0, 1].
    transforms.Normalize(mean=(0.5,), std=(0.5,)), # 전처리 [0,1] -> [-1, 1]
          # 평균이 0에 가까워져서 신경망 학습 (ReLU, tanh) 에서 학습효율이 높아짐
          # 요즘 사전학습모델들 이 이미 [-1, 1] 구간으로 학습된것이 많음.
          # channel 이 3개면. mean=[0.4914, 0.4822, 0.4465],
          #                  std=[0.2023, 0.1994, 0.2010]

])


# 위와 같이 할수도 있고,
# 아래와 같이 전처리 정보를 별도로 준비해서 적용할수도 있다
```

```python
# 전처리 정의
transform_config = {
    "resize": (28, 28),
    # ↓각 채널별로 mean, std 값 준비
    "mean": [0.5],
    "std": [0.5],
    # 만약 RGB 3개 채널이면 아래와 같이 해야 한다.
    # "mean": [0.5, 0.5, 0.5],
    # "std": [0.5, 0.5, 0.5]
}

# transforms 정의 (훈련/추론 동일하게 사용해야 한다)
transform = transforms.Compose([
    transforms.Resize(transform_config["resize"]),
    transforms.ToTensor(),
    transforms.Normalize(mean=transform_config["mean"], std=transform_config["std"])
])
```

## 왜 mean=0.5, std=0.5 인가?

`transforms.Normalize(mean=0.5, std=0.5)` 같은 전처리를 많이 보게 되는데 이유는 다음과 같습니다.

---

### 1. 데이터 정규화의 기본 개념

신경망 학습에서 입력 데이터를 **특정 분포(보통 평균 0, 분산 1)** 로 맞추면 학습이 더 안정적이고 빠르게 진행됩니다.

이유는:

* 각 입력 특성(feature)의 스케일이 비슷해져서 **가중치 업데이트가 균형** 있게 일어남
* 활성화 함수(특히 `sigmoid`, `tanh`)의 **비선형 구간**을 잘 활용할 수 있음
* 경사 하강법에서 **기울기 소실/폭발** 위험을 줄여줌

---

### 2. 0.5, 0.5를 쓰는 구체적 이유

일반적으로 이미지 데이터는 `transforms.ToTensor()` 후 `[0,1]` 범위로 바뀝니다.
여기서 `Normalize(mean=0.5, std=0.5)`를 적용하면:

$x' = \frac{x - 0.5}{0.5} = 2x - 1$

즉, 입력 값 범위가:
$[0, 1] \to [-1, 1]$ 으로 변환됩니다.

---

### 3. 왜 [-1, 1]로 맞추는가?

* **대칭적인 분포**: 평균이 0에 가까워져서 신경망(특히 `tanh`, `ReLU` 류 활성화 함수)이 학습하기 쉬워짐
* **빠른 수렴**: 가중치 초기화와 잘 어울려 학습 속도가 빨라짐
* **실무적 편의**: 많은 pretrained 모델(CNN 등)이 이미 [-1,1] 구간을 가정하고 학습되어 있음

---

### 4. 다른 경우

* CIFAR-10, ImageNet 같은 경우는 실제 데이터셋의 **채널별 평균/표준편차**를 구해서 정규화 (`mean=[0.491, 0.482, 0.447], std=[0.247, 0.243, 0.262]` 등)
* 하지만 간단한 실험이나 GAN 학습에서는 `0.5, 0.5`가 가장 흔히 쓰임 → 픽셀을 단순히 `[-1, 1]`로 스케일링하기 위함

---

✅ 정리

`mean=0.5, std=0.5` 전처리를 하는 이유

 **[0,1] 범위의 이미지를 [-1,1]로 스케일링해서 신경망 학습을 안정적이고 효율적으로 만들기 위해서!**

# 모델 정의 & 생성 ✅

```python
# import
from torch import nn
```

```python
# model 정의
class MyCNNModel(nn.Module):

  def __init__(self):
    super().__init__()

    # 모델을 구성하는 레이어들 정의

    # 입력: 1채널  28x28 이미지
    # conv1: (1->32채널, 3x3커널), 출력 28x28
    self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, padding=1)

    # 입력: 32채널  14x14 이미지
    # conv2: (32->64채널, 3x3커널), 출력 14x14
    self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1)


    # pooling: 크기 절반 (ex: 28x28 -> 14x14)
    self.pooling = nn.MaxPool2d(kernel_size=2, stride=2)

    # 입력:(3136,) -> 출력: (256,)
    self.fc1 = nn.Linear(64 * 7 * 7, 256)

    # 입력: (256,) -> 출력: (10,)
    self.fc2 = nn.Linear(256, 10)

    # Dropout
    self.dropout25 = nn.Dropout(p=0.25)
    self.dropout50 = nn.Dropout(p=0.5)


  # 모델 입력 data 가 주어졌을때 순방향 전파 로직 정의.
  def forward(self, data):  # <- data: 모델 입력
    # [Feature Extraction]
    data = self.conv1(data)
    data = torch.relu(data)
    data = self.pooling(data)
    data = self.dropout25(data)

    data = self.conv2(data)
    data = torch.relu(data)
    data = self.pooling(data)
    data = self.dropout25(data)


    # [Classification]
    data = data.view(-1, 64 * 7 * 7)   # 1차원으로 펼치기

    data = self.fc1(data)
    data = torch.relu(data)
    data = self.dropout50(data)

    logits = self.fc2(data)

    return logits  # 모델의 출력

```

```python
# GPU 로 모델 이동
model = MyCNNModel().to(DEVICE)
```

# 손실함수 & optimizer ✅

```python
# 손실 함수: CrossEntropyLoss
#     softmax + negative log likelihood 결합.  <- softmax 함수가 포함되어 있다
#     다중 분류 문제에서 자주 사용.

loss_function = nn.CrossEntropyLoss()

# Optimizer: Adam , 학습률 0.001 !
optimizer = torch.optim.Adam(model.parameters(), lr= 1e-3)
```

# train, eval, test 함수 정의 ✅

```python
# train 함수
def model_train(dataloader, model, loss_function, optimizer):

  model.train()  # 학습모드 (Dropout 활성화)

  train_loss_sum = 0  # 각 batch 마다 loss 값 누적합산.
  train_total = 0    # 각 batch 에서 학습된 데이터 개수누적
  train_correct = 0  # 각 batch 마다 맞춘 개수 누적

  total_train_batch = len(dataloader) # 배치의 개수

  for images, labels  in dataloader:  # 각 배치 마다

    x_train = images.to(DEVICE)
    y_train = labels.to(DEVICE)

    outputs = model(x_train)  # 모델입력 -> forward() 가 호출.
    loss = loss_function(outputs, y_train) # 손실(loss) 계산

    optimizer.zero_grad() # 역전파 하기 전에 기울기 초기화
    loss.backward()  # 역전파 (backprogation)  기울기 계산
    optimizer.step()  # 가중치 업데이트 (가중치 최적화)

    train_loss_sum += loss.item()   # 매 batch 마다의 loss 값
    train_total += y_train.size(0)

    # 이번 batch 에서 모델이 맞춘 샘플 수를 계산해서 누적
    # outputs 의 형태는 (batch, 10)
    # torch.argmax(outputs, dim=1)  # -> 결과 형태 (batch,) 정수인덱스 (0 ~ 9)
    # torch.argmax(outputs, dim=1) == y_train  # 결과 형태 (batch,) True, False
    # (torch.argmax(outputs, dim=1) == y_train).sum().item()  # 맞춘 개수
    train_correct += (torch.argmax(outputs, dim=1) == y_train).sum().item()

  # 평균 loss, accuracy 계산
  train_avg_loss = train_loss_sum / total_train_batch
  train_avg_accuracy = 100 * train_correct / train_total

  return (train_avg_loss, train_avg_accuracy)
```

```python
# evaluate 함수
def model_evaluate(dataloader, model, loss_function): # 🔨평가함수에는 optimizer 필요없슴.

  model.eval()  # 평가 모드 (Dropout 비활성화)

  with torch.no_grad():  # 기울기 계산 비활성화

      val_loss_sum = val_correct = val_total = 0

      total_val_batch = len(dataloader)

      for images, labels in dataloader:

          x_val = images.to(DEVICE)
          y_val = labels.to(DEVICE)

          outputs = model(x_val)
          loss = loss_function(outputs, y_val)

          val_loss_sum += loss.item()

          val_total += y_val.size(0)
          val_correct += ((torch.argmax(outputs, 1)==y_val)).sum().item()

      val_avg_loss = val_loss_sum / total_val_batch
      val_avg_accuracy = 100*val_correct / val_total

  return (val_avg_loss, val_avg_accuracy)


```

```python
# test 함수
# 검증 함수와 비슷하지만, 최종 성능(accuracy, loss)만 출력.
def model_test(dataloader, model):

  model.eval()

  with torch.no_grad():

      test_loss_sum = test_correct = test_total = 0

      total_test_batch = len(dataloader)

      for images, labels in dataloader:

          x_test = images.to(DEVICE)
          y_test = labels.to(DEVICE)

          outputs = model(x_test)
          loss = loss_function(outputs, y_test)

          test_loss_sum += loss.item()

          test_total += y_test.size(0)
          test_correct += ((torch.argmax(outputs, 1)==y_test)).sum().item()

      test_avg_loss = test_loss_sum / total_test_batch
      test_avg_accuracy = 100*test_correct / test_total

      print('accuracy:', test_avg_accuracy)
      print('loss:', test_avg_loss)


```

# 학습 함수 정의 ✅

```python
# epoch 진행, 순전파-loss계산 - 역전파 - 파라미터 업데이트,
# loss 기록, 진행시간 체크
def train(model, epochs,
          train_dataset_loader, validation_dataset_loader,
          loss_function, optimizer):

  train_loss_list = []
  train_accuracy_list = []

  val_loss_list = []
  val_accuracy_list = []

  start_time = datetime.now()

  for epoch in range(epochs):

    # ==== model train ====
    train_avg_loss, train_avg_accuracy = model_train(train_dataset_loader, model, loss_function, optimizer)
    train_loss_list.append(train_avg_loss)
    train_accuracy_list.append(train_avg_accuracy)
    #=============================================

    # ==== model evaluate ===
    val_avg_loss, val_avg_accuracy = model_evaluate(validation_dataset_loader, model, loss_function) # 🔨평가함수에는 optimizer 필요없슴.

    val_loss_list.append(val_avg_loss)
    val_accuracy_list.append(val_avg_accuracy)
    #=============================================

    # 결과 출력
    print('epoch:', '%02d' % (epoch + 1),
          'train loss =', '{:.4f}'.format(train_avg_loss), 'train acc =', '{:.4f}'.format(train_avg_accuracy),
          'val loss =', '{:.4f}'.format(val_avg_loss), 'val acc =', '{:.4f}'.format(val_avg_accuracy))


  end_time = datetime.now()
  print('elapsed time => ', end_time-start_time)

  return train_loss_list,train_accuracy_list, val_loss_list, val_accuracy_list

```
