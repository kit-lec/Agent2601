# 손익계산서

```python
# 연간 손익계산서 (rows = 항목,  columns=연도)
df = ticker.income_stmt
df
```

```python
df_t = df.T
df_t
```

```python
df_t.index = df_t.index.astype(str)  # Timestamp -> str  (JSON 변환을 위해)

df_t
```

```python
"""
이런 데이터를 만들어 볼거다.
[{'fiscalDateEnding': '2025-12-31',
  'Tax Effect Of Unusual Items': -133380000.0,
  'Tax Rate For Calcs': 0.27,
  'Normalized EBITDA': 12258000000.0,
  'Total Unusual Items': -494000000.0,
...
  'Total Revenue': 94827000000.0,
  'Operating Revenue': 94827000000.0},
 {'fiscalDateEnding': '2024-12-31',
  'Tax Effect Of Unusual Items': -136800000.0,
  'Tax Rate For Calcs': 0.2,
  'Normalized EBITDA': 15392000000.0,
  'Total Unusual Items': -684000000.0,
"""
None
```

```python
records = []

for year, row in df_t.iterrows():
    # print(year, type(row))
    record = {"fiscalDateEnding": year}

    for key, value in row.items():
        # NaN 값은 None 으로 정리
        if value is None or (isinstance(value, float) and value != value):
            record[str(key)] = None
        else:
            record[str(key)] = value

    records.append(record)

records
```

# 주가 정보

```python
# 주봉(weekly) 데이터, 최근 6개월
hist = ticker.history(period='6mo', interval='1wk', auto_adjust=False)
hist
```

```python
# 최근 12주만 가져오기 (토큰 절약)
hist = hist.tail(12)
hist
```

```python
"""
[{'date': '2026-03-02',
  'open': 390.6,
  'high': 408.62,
  'low': 385.39,
  'close': 396.73,
  'volume': 301991600},
 {'date': '2026-03-09',
  'open': 390.05,
  'high': 416.38,
  ...
"""
None
```

```python
weekly_bars = []

for ts, row in hist.iterrows():
    weekly_bars.append({
        "date": ts.strftime("%Y-%m-%d"),
        "open": round(float(row["Open"]), 4),
        "high": round(float(row["High"]), 4),
        "low": round(float(row["Low"]), 4),
        "close": round(float(row["Close"]), 4),
        "volume": int(row["Volume"]),        
    })

weekly_bars
```

## 간단한 성과 요약

```python
first_close = weekly_bars[0]['close']
last_close = weekly_bars[-1]['close']
change_pct = ((last_close - first_close) / first_close) * 100.0

change_pct
```

```python
result = {
    "symbol": symbol,
    "interval": "1 week",
    "weeks_returned": len(weekly_bars),
    "summary": {
        "first_close": first_close,
        "last_close": last_close,
        "period_change_pct": round(change_pct, 2),
        "period_high": round(float(hist["High"].max()), 4),
        "period_low": round(float(hist["Low"].min()), 4),
    },
    "weekly_bars": weekly_bars,    
}

result
```
