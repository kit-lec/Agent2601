## 리턴하는 컬럼

한국 금융감독원 전자공시시스템(DART)의 Open API 데이터를 기반으로 하는 **DartFinanceReader** (주로 `dart-fss` 나 `OpenDartReader` 등에서 재무제표를 파싱할 때 사용)가 반환하는 DataFrame 컬럼들은 오픈DART의 **'재무제표 주요계정'** 및 **'단일회사 전체재무제표'** API 표준 항목들입니다.

---

### 1. 공시 및 기업 기본 정보

* **`rcept_no`** (접수번호): 해당 재무제표가 제출된 공시 보고서의 고유 번호 (14자리)입니다.
* **`reprt_code`** (보고서 코드): 공시 보고서의 종류를 나타내는 코드입니다.
    * `11013`: 1분기 보고서
    * `11012`: 반기 보고서
    * `11014`: 3분기 보고서
    * `11011`: 사업 보고서 (연말 결산)


* **`bsns_year`** (사업연도): 재무제표가 속한 기준 연도 (예: `2025`)입니다.
* **`corp_code`** (고유번호): DART에서 기업을 식별하기 위해 부여한 8자리 고유 고유번호입니다. (상장사 주식종목코드와 다름)
* **`stock_code`** (종목코드): 한국거래소(KRX) 상장 주식 종목 코드 (6자리)입니다. 비상장사인 경우 비어있을 수 있습니다.

---

### 2. 재무제표 구분 정보

* **`fs_div`** (재무제표구분): 연결재무제표와 별도(개별)재무제표를 구분하는 코드입니다.
    * `CFS`: 연결재무제표 (Consolidated Financial Statements)
    * `OFS`: 별도/개별재무제표 (Separate Financial Statements)


* **`fs_nm`** (재무제표명): 위 구분을 한글 명칭(`연결재무제표` 또는 `별도재무제표`)으로 표시한 것입니다.
* **`sj_div`** (재무제표종류 구분): 재무제표의 종류를 나타내는 코드입니다.
    * `BS`: 재무상태표 (Balance Sheet)
    * `IS`: 손익계산서 (Income Statement)
    * `CIS`: 포괄손익계산서 (Comprehensive Income Statement)
    * `CF`: 현금흐름표 (Cash Flow Statement)
    * `SCE`: 자본변동표 (Statement of Changes in Equity)


* **`sj_nm`** (재무제표종류명): 위 구분을 한글 명칭(`재무상태표`, `손익계산서` 등)으로 표시한 것입니다.
* **`account_nm`** (계정과목명): 재무제표에 표시된 계정의 이름 (예: `유동자산`, `매출액`, `당기순이익` 등)입니다.

---

### 3. 기간 및 금액 정보 (3개년도 비교)

DART 재무제표는 기본적으로 당기, 전기, 전전기 3개년 데이터를 함께 제공합니다.

| 구분 | 컬럼명 | 의미 | 설명 |
| --- | --- | --- | --- |
| **당기 (이번 분기/개념)** | **`thstrm_nm`**<br><br>**`thstrm_dt`**<br><br>**`thstrm_amount`** | 당기명<br><br>당기 일<br><br>당기 금액 | 예: 제 25 기<br><br>예: 2025.12.31 현재 (또는 기간)<br><br>해당 계정과목의 **현재 분기/연도 금액** |
| **전기 (직전 연도)** | **`frmtrm_nm`**<br><br>**`frmtrm_dt`**<br><br>**`frmtrm_amount`** | 전기명<br><br>전기 일자<br><br>전기 금액 | 예: 제 24 기<br><br>예: 2024.12.31 현재<br><br>해당 계정과목의 **직전 연도 금액** |
| **전전기 (2년 전 연도)** | **`bfefrmtrm_nm`**<br><br>**`bfefrmtrm_dt`**<br><br>**`bfefrmtrm_amount`** | 전전기명<br><br>전전기 일자<br><br>전전기 금액 | 예: 제 23 기<br><br>예: 2023.12.31 현재<br><br>해당 계정과목의 **2년 전 연도 금액** |

> 💡 **참고:** `thstrm` = This Term (당기), `frmtrm` = Former Term (전기), `bfefrmtrm` = Before Former Term (전전기)의 약어입니다.

---

### 4. 기타 정렬 및 통화 정보

* **`ord`** (정렬순서): 재무제표 상에서 해당 계정과목이 나오는 출력 순서입니다. 데이터를 원본 재무제표 순서대로 정렬할 때 유용합니다.
* **`currency`** (통화): 금액의 화폐 단위 (예: `KRW` 등)입니다.

---

## 🔵함수 정의

````python
def get_financial_statement(
    corp_code,
    date,
    report_code,
    sj_div,
):
    df = dart.finstate(corp_code, date, report_code)
    # 연결재무재표
    filtered_df = df[(df['fs_div'] == 'CFS') & (df['sj_div'] == sj_div)]
    if filtered_df.empty:
        # 혹시 없다면 OFS
        filtered_df = df[(df['fs_div'] == 'OFS') & (df['sj_div'] == sj_div)]

    filtered_df = filtered_df[['corp_code', 'bsns_year', 'reprt_code', 'account_nm', 'thstrm_amount']]
    return filtered_df
````

````python
inputs = {
    'corp_code': '00126380',
    'date': '2025',
    'report_code': '11013',
    'sj_div': 'BS'
}

get_financial_statement(**inputs)
````

## 🔵 함수정의

````python
EVENT_CODES = [
    '부도발생', '영업정지', '회생절차', '해산사유', '유상증자', '무상증자', '유무상증자', '감자',
    '관리절차개시', '소송', '해외상장결정', '해외상장폐지결정', '해외상장', '해외상장폐지',
    '전환사채발행', '신주인수권부사채발행', '교환사채발행', '관리절차중단', '조건부자본증권발행',
    '자산양수도', '타법인증권양도', '유형자산양도', '유형자산양수', '타법인증권양수', '영업양도',
    '영업양수', '자기주식취득신탁계약해지', '자기주식취득신탁계약체결', '자기주식처분', '자기주식취득',
    '주식교환', '회사분할합병', '회사분할', '회사합병', '사채권양수', '사채권양도결정'
]
````

````python
def get_major_event_report(
    corp_code,
    event,
    date
):
    if event not in EVENT_CODES:
        return {"error": f"event must be one of: {EVENT_CODES}"}

    result = dart.event(corp_code, event, date)

    if isinstance(result, pd.DataFrame) and result.empty:
        return {'message': "No data found for the given parameters."}

    return result

    
````
