# AI Agent
AI 에이전트(AI Agent)는 주어진 목표를 달성하기 위해 주변 환경을 인지하고, 스스로 판단하여 행동을 수행하는 자율적인 인공지능 시스템을 말합니다.

단순히 사용자의 질문에 답을 생성하는 '챗봇(Chatbot)'이나 일정한 규칙에 따라 동작하는 '자동화 프로그램'을 넘어, **목표를 달성하기 위한 최선의 경로를 스스로 계획하고 실행**한다는 점에서 차세대 AI 기술로 주목받고 있습니다.

AI 에이전트의 핵심 메커니즘을 쉽게 이해하기 위해 구조를 살펴보면 다음과 같습니다.

---

## 1. AI 에이전트의 4대 핵심 요소

AI 에이전트가 자율적으로 작동할 수 있는 것은 대형언어모델(LLM)을 '두뇌'로 삼고 아래의 4가지 핵심 능력을 결합했기 때문입니다.

* **목표 및 계획 수립 (Planning):** 거대한 목표가 주어지면 이를 해결 가능한 작은 단위의 작업(Sub-tasks)으로 쪼갭니다. 과거의 실수를 돌아보고 계획을 수정하는 '자기 반성(Self-reflection)' 능력도 포함됩니다.
* **기억장치 (Memory):**
    * *단기 기억:* 현재 진행 중인 대화나 작업 맥락을 기억합니다.
    * *장기 기억:* 외부 데이터베이스나 벡터 DB를 활용해 오랜 시간이 지난 후에도 필요한 정보를 찾아내고 유지합니다.


* **도구 활용 (Tools):** AI 자체의 한계를 극복하기 위해 외부 도구를 사용합니다. 계산기 활용, 웹 브라우징, API 호출, 데이터베이스 조회 등이 이에 해당합니다.
* **행동 (Action):** 계획과 도구를 바탕으로 이메일 발송, 결제, 코드 실행 등 실제 환경이나 시스템에 변화를 주는 결과물을 만들어냅니다.

---

## 2. 챗봇(Chatbot) vs AI 에이전트(Agent)

| 구분 | 일반 AI 챗봇 (예: 기본 ChatGPT, Gemini) | AI 에이전트 (AI Agent) |
| --- | --- | --- |
| **작동 방식** | 사용자의 질문(프롬프트)에 수동적으로 답변 | 목표를 주면 스스로 판단하여 능동적으로 행동 |
| **수행 단계** | 1문 1답 형태의 단발성 상호작용 | 목표 달성을 위해 멀티 스텝(Multi-step) 작업 수행 |
| **도구 활용** | 텍스트 생성 및 내장된 기능 위주 활용 | 브라우저 제어, API 연동 등 외부 도구 자율 제어 |
| **예시** | "서울 맛집 3곳 추천해줘" $\rightarrow$ 목록 작성 | "내 일정과 예산에 맞춰 서울 여행 계획을 짜고 숙소까지 예약해줘" $\rightarrow$ 대리 수행 |

---

## 3. 대표적인 활용 시나리오

AI 에이전트는 이미 다양한 영역에서 실험 단계를 넘어 실무에 적용되고 있습니다.

* **업무 자동화 및 비서:** 사용자의 메일함을 관리하고, 중요도를 판단해 답장 초안을 작성하거나, 회의 일정을 조율하고 캘린더에 자동으로 등록합니다.
* **웹 브라우저 제어 및 예약:** 인간 대신 웹사이트에 접속해 로그인하고, 원하는 조건의 항공권이나 식당을 찾아 최종 결제 직전 단계까지 자율적으로 진행합니다.
* **소프트웨어 개발 (AI 개발자):** 코드의 버그를 찾아달라고 요청하면, 스스로 전체 소스코드를 분석하고 수정한 뒤 테스트까지 완료하여 패치 코드를 제안합니다.
* **디바이스 제어 (스크린 자동화):** 스마트폰이나 PC의 화면(UI)을 직접 인식하고 터치/클릭하여 배달 앱에서 음식을 주문하거나 택시를 호출하는 등의 작업을 수행합니다.

> 💡 **요약하자면**
> 기존의 AI가 질문에 답을 주는 **'지식 가이드'** 였다면, AI 에이전트는 사용자를 대신해 복잡한 업무를 끝마쳐주는 **'자율적인 대리인(Agent)'** 이라고 볼 수 있습니다.

# 1.그래프(Graph) 자료구조의 이해

랭그래프를 제대로 이해하려면 먼저 그래프 Graph 자료구조에 대한 기본 개념을 알아야 합니다. 

## 노드 & 에지

그래프(Graph)는 **노드(Node)** 와 **에지(Edge)** 로 구성된 자료구조로, 객체 간의 **관계**를 표현하는 데 매우 효과적입니다. 

※자료구조에선 노드를 버텍스(Vertex), 에지를 아크(Arc) 라고도 표기하기도 함.

![](https://dz2cdn1.dzone.com/storage/temp/13817670-figure-1.png)


![](https://www.computersciencebytes.com/wp-content/uploads/2017/01/vertices_edges.png)

## Directed vs. Undirected
그래프는 방향성에 따라 방향 그래프 Directed Graph 와 무방향 그래프 Undirected Graph 로 나뉩니다. 

![](https://www.chhitizbuchasia.com/static/9a65c24a1cc442a1c313eee8f8796494/f73a1/graph-direction.png)

**랭그래프는 방향 그래프를 기반**으로 합니다. 이는 작업의 흐름이 특정 방향을 가지기 때문입니다. 
예를 들어 A 작업이 완료된 후 B 작업이 실행되는 것을 그래프로 명확하게 표현할 수 있습니다.

## Cyclic vs. Acyclic 
![](https://i.imgur.com/2z9J2E5.png)
그래프의 또 다른 중요한 특성은 순환(cycle)의 존재 여부입니다. 
순환이 없는 그래프를 **DAG(Directed Acyclic Graph)** 라고 하며, 많은 워크플로 시스템이 DAG만을 이용합니다. 

랭그래프는 __순환이 있는 그래프도 사용할 수 있습니다__. 이를 통해 조건이 만족될 때까지 특정 작업을 반복하거나, 
이전 단계로 돌아가는 등의 복잡한 패턴을 구현할 수 있습니다.

# 2.랭그래프의 핵심 개념
랭그래프에서 가장 중요한 개념은 **상태(state), 노드(node), 에지(edge)** 다

## 상태(state)
**상태**는 그래프 실행 과정에서 지속적으로 유지되는 데이터로 각 노드가 실행될 때마다 읽고 쓸 수 있습니다. 

이는 단순한 변수 전달을 넘어 전체 워크플로의 컨텍스트를 관리하는 **중앙 저장소 역할**을 합니다. 

상태는 일반적으로 **TypedDict**나 **Pydantic 모델**로 정의되어 타입 안정성을 보장합니다.

![](https://cdn.prod.website-files.com/65b8cd72835ceeacd4449a53/66dba5f465f6e9a2482ad935_simple-graph1.png)

## 노드(node)
'노드' 는 그래프의 기본 실행 단위입니다.  각 노드는 특정 작업을 수행하는 **함수**나 **에이전트**를 나타내며, 다음과 같은 특징을 가집니다.

- 현재 상태를 **입력**으로 받음
- 특정 **작업을 수행**(LLM 호출, 데이터 처리, 외부 API 호출 등)
- **업데이트된 상태를 반환(리턴)**

노드는 순수함수일 수도 있고, 부수 효과side Effects를 가질 수도 있습니다. 예를 들어 DB에 쓰기 작업을 하거나 외부 API를 호출할 수도 있습니다.

## 에지(edge)
'에지'는 노드 간의 연결을 정의하며, 실행 흐름을 제어합니다. 랭그래프는 두 가지 유형의 에지를 지원합니다.

1. **일반 에지** : 항상 같은 경로로 진행

2. **조건부 에지**: 상태에 따라 다른 노드로 분기

조건부 에지는 동적 라우팅을 가능하게 하여, 런타임에 실행 경로를 결정할 수 있습니다. 
이를 사용하여 AI 에이전트가 상황에 따라 다른 전략을 선택하게 하거나, 오류 처리 로직을 구현하는 데 유용합니다.

* 일반에지는 실선으로 표현하고, 조건부 에지는 점선으로 표기하기도 한다

# 3. Hello LangGraph

```python
# [시작] --> [greeting] --> [Processing] --> [종료]
```

```python
from typing import Dict, Any
from langgraph.graph import StateGraph, START, END
from pydantic import BaseModel, Field
from IPython.display import Image, display
```

## '상태' 정의

```python
# 워크플로우 단계 정의

class WorkflowStep:
    GREETING = 'GREETING'   # 인사말 생성 단계
    PROCESSING = 'PROCESSING'   # 생성된 인사말을 처리하고 메세지를 생성하는 단계
```

```python
# 그래프 상태 정의
# 그래프의 상태는 각 노드마다 참조가 가능한 전체워크플로 상의 컨텍스트 라고 볼수 있다.
class GraphState(BaseModel):
    name: str = Field(default="", description="사용자 이름")
    greeting: str = Field(default="", description="생성된 인사말")
    processed_message: str = Field(default="", description="처리된 최종 메시지")    
```

## '노드' 정의

```python
# 첫번째 노드 함수
def generate_greeting(state: GraphState) -> Dict[str, Any]:
    name = state.name or '아무개'
    greeting = f"안녕하세요, {name} 님!"
    print(f"[generate_greeting] 인사말 생성: {greeting}")
    return {"greeting": greeting}  # state 의 "greeting" 값 업데이트
```

```python
# 두번째 노드 함수 : 인사말 처리하고 최종 메세지 생성
def process_message(state: GraphState) -> Dict[str, Any]:
    greeting = state.greeting
    processed_message = f"{greeting} LangGraph에 오신 것을 환영합니다!"
    print(f"[process_message] 최종 메세지: {processed_message}")
    return {"processed_message": processed_message}
    
```

## '그래프(워크플로)' 정의

```python
# 그래프 정의 함수
def create_hello_langgraph():
    # '상태'를 가진 워크플로(그래프) 객체 생성
    # StateGraph(상태)
    workflow = StateGraph(GraphState)

    # 워크플로 객체에 노드(들) 추가
    # add_node(노드이름, 노드함수)
    workflow.add_node(WorkflowStep.GREETING, generate_greeting)
    workflow.add_node(WorkflowStep.PROCESSING, process_message)

    # 엣지 추가 (노드 간 연결)
    # add_edge(노드1, 노드2)  노드1 -> 노드2 엣지 형성
    workflow.add_edge(START, WorkflowStep.GREETING)   # START : 시작점
    workflow.add_edge(WorkflowStep.GREETING, WorkflowStep.PROCESSING)
    workflow.add_edge(WorkflowStep.PROCESSING, END)   # END: 그래프 종료지점


    # 그래프 컴파일 -> 실행가능한(Runnable) 그래프 객체 생성  ( .invoke() 로 호출 가능! )
    app = workflow.compile()

    return app
```

## 그래프 생성 및 실행

```python
def main():
    print("Hello LangGraph\n")
    # 그래프 생성
    app = create_hello_langgraph()

    # 초기 상태
    initial_state = GraphState(name="감귤", greeting="", processed_message="")
    print('🟡 초기 상태:', initial_state.model_dump())

    # 그래프 실행
    print("🟩 그래프 실행 시작")
    final_state = app.invoke(initial_state)  # 초기상태 전달하여 그래프 호출. 리턴값은 상태의 dict
    print("🟩 그래프 실행 종료")
    print('🟡 최종 상태:', final_state, type(final_state))

    print(f"결과메시지: {final_state['processed_message']}")


    # 그래프 이미지 생성
    result = app.get_graph().draw_mermaid_png()
    with open('./hello_langgraph.png', 'wb') as f:
        f.write(result)

main()    
```
