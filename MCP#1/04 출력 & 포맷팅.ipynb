# 기본출력함수 print()

```python
# 파이썬의 가장 기본적인 출력 함수 print()

# print( 출력내용1, 출력내용2, 출력내용3, ...)
#  위와 같이 print() 함수의 parameter 로 출력할 내용들을 기입하면 된다

# print() 함수는 출력한후 자동적으로 "줄바꿈" 됩니다 (디폴트)
```

```python
print(10, 20, 30)
print()
print('hello')
print()
print(3.14)
```

## end=
출력이 끝난뒤 마지막에 출력하는 문자열

```python
print(10, 20, 30, end='--')
print('hello')
print(3.14)
```

```python
print(10, 20, 30, end='\n')  #'\n'  줄바꿈 문자
print('hello', end='***')
print(3.14)
```

## sep=
separator.  출력할 값들간 구분문자열

```python
print(10, 20, 30, sep=' ')
```

```python
print(10, 20, 30, sep='**')
```

# 문자열 포맷팅 (formatting)
- 서식화된 문자열

## 방법1 % 연산자 사용

```python
# 서식지정자(format specifier) 를 포함한 문자열과,
#  각 '서식문자'에 대응하는 데이터들을 연결하여 문자열 완성

# 구문)
#    "서식문자포함문자열" % ( 데이터 튜플 )
```

```python
a = "Hello %s"

a
```

```python
a % ("파이썬")
```

```python
a % ("MCP")
```

```python
"내 이름은 %s, 나이는 %d살" % ("홍길동", 20)
```

```python
# format specifier (서식 지정자)
#  https://docs.python.org/2/library/string.html

# %d   십진수 정수로 출력
# %f   실수로 출력
# %s   문자열로 출력
# %%   '%' 문자 자체
```

```python
"%d %f" % (10, 10)
```

```python
import math
PI = math.pi
PI
```

```python
"%d %f %.1f %.2f %.3f" % (PI, PI, PI, PI, PI)
```

## 방법2 format() 사용

```python
myFmt3 = "My name is {}, I'm {} years old".format("박수진", 10)

myFmt3
```

```python
myFmt3 = "My name is {name}, I'm {age} years old".format(name="박수진", age=10)

myFmt3
```

```python
myFmt3 = "My name is {name}, I'm {age} years old".format(age=10, name="박수진")

myFmt3
```

```python
"PI = {}".format(PI)
```

```python
"PI = {:.3f}".format(PI)
```

```python
"price = {}".format(1234567890)
```

```python
"price = {:,}".format(1234567890)
```

## 방법3 f-string 사용

```python
lang = 'Python'
author = 'Guido van Rossum'
```

```python
'Language: {}, Author: {}'.format(lang, author)
```

```python
f'Language: {lang}, Author: {author}'
```

```python
f'원주율은 {PI:.2f} 입니다'
```

# 이스케이프 문자 (escape character)

```python
# '문자열' 내에서 특수한 문자 출력할때 사용
# \와 조합하여 정의됨

# 많이 사용되는 이스케이프 코드
#    \n : 줄바꿈
#    \t : 탭
#    \\ : 역슬래시
#    \' : 홀따옴표
#    \" : 쌍따옴표
```

```python
print("1.Hello Python")
print("2.Hello \nPython")
print("3.He\tllo \tPython")
print("4.Hello\" \\Python")
```

# Raw String  (r-string)
문자열 내부의 모~든 특수기능 제거한 문자열

```python
# 윈도우 는 file separator 문자가 back slash
path = "C:\Program Files\Microsoft\011_python\times"
print(path)
```

```python
path = "C:\\Program Files\\Microsoft\\011_python\\times"
print(path)
```

```python
path = r"C:\Program Files\Microsoft\011_python\times"
print(path)
```
