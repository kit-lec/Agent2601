```python
#  아래와 같은 경우를 생각해보자.

def func_sum(data):
    result = 0
    for i in data:
        result = result + i
    return result
```

# 예외(오류) 처리 : try ~ except

```python
# 물론 사전에 if 등의 조건식을 활용해서 오류 를 사전에 예방은 가능하나

a = 10
b = 0
if b != 0:
    print(a / b)
else:
    print('0 으로 나눌수 없습니다')

# 수~~~많은 예외상황을 매번 if 조건문으로 처리하면
# 코드가 매우 난잡해진다. --> 유지보수 힘들어짐.

# 그래서, 예외처리 관련하여
# '코드수행블럭'  와 '예외처리블럭' 구분하여 예외를 다루는게 바람직.
```

```python
# 파이썬 에서는
# try ~ except 문으로 오류를 다룹니다

# 구문]
# try:
#     코드수행블럭
#     ...
# except [발생 오류[as 오류 메시지 변수]]:
#     예외처리블럭
#     ...

#  try 블록 수행 중 오류가 발생하면 except 블록이 수행된다.
# 하지만 try블록에서 오류가 발생하지 않는다면 except 블록은 수행되지 않는다.
```

## 1. try ~ except

```python
num = 0

try:
  print('1-1 try 시작')
  result = 10 / num  # try 블럭 안에서 예외 발생하면 except: 넘어감
  print('1-2 try 종료 result=', result)
except:  # '어떠한 에러가' 발생해도 여기서 처리(handling)
  print('2.오류발생')

print('4. 프로그램 종료')
```

 ## 2.  try: ~ except 오류:

```python
num = 0
# num = "0"

try:
  print('1-1 try 시작')
  result = 10 / num # except 에서 처리되지 않은 에러 => 프로그램 종료.
  print('1-2 try 종료 result=', result)
# except ZeroDivisionError:   # except 에는 처리하고자 하는 에러 명시 가능
except TypeError:
  print('2.오류발생')

print('4. 프로그램 종료')
```

```python
# num = 0
num = "0"

try:
  print('1-1 try 시작')
  result = 10 / num
  print('1-2 try 종료 result=', result)
# ↓ except: 는 여러개 명시 가능!
except ZeroDivisionError:
  print('2-1 ZeroDivisionError 발생')
except TypeError:
  print('2-2.TypeError 발생')

print('4. 프로그램 종료')

```

## 3. try ~ except 오류 as 변수:

```python
# num = 0
num = "0"

try:
  print('1-1 try 시작')
  result = 10 / num
  print('1-2 try 종료 result=', result)
except ZeroDivisionError:
  print('2-1 ZeroDivisionError 발생')
except TypeError as e:  # 발생한 예외객체를 변수 e 에 받아서 처리
  print('2-2.TypeError 발생')
  print('🟨', e)
  print('🟦', e.args)  # tuple <- 예외 관련된 메세지, 추가 정보등이 담겨있다.

print('4. 프로그램 종료')
```

## 4. finally
예외가 발생했든 아니든 반드시 수행해야 하는 코드들은 finally  에 작성

```python
f = open('foo.txt', 'w')

a = 10
a = a + bb

print('1. 파일 처리 완료')
f.close()  # <- 꼭 반드시 실행되어야 하는 코드인데 실행이 안된다!!!?
```

```python
# finally 사용

try:
  print('1-1 try 시작')
  f = open('foo.txt', 'w')
  a = 10
  # a = a + bb
  print('1-2 try 종료')
except:
  print('2. 오류 발생')
finally:
  print('3. finally')
  f.close()



print('4. 프로그램 종료')
```

## except 오류, 오류 ....

```python
try:
  print('1-1. try: 시작')
  a = ['alpha', 'beta']
  print(a[1])
  b = 4 / 0
  print("1-2. try 블럭 종료")
except IndexError as e:
  print("2-1. IndexError 발생", e.args)
except ZeroDivisionError as e:
  print("2-2. ZeroDivisionError 발생", e.args)

print('4 프로그램 종료')
```

```python
try:
  print('1-1. try: 시작')
  a = ['alpha', 'beta']
  print(a[1])
  b = 4 / 0
  print("1-2. try 블럭 종료")
except (IndexError, ZeroDivisionError) as e:
  print("2. Error발생", e.args, e.__class__.__name__, type(e))

print('4 프로그램 종료')

```

# 예외 발생: raise
강제로 오류 발생 시키기

```python
# 프로그래밍을 하다 보면 종종 오류를 일부러 발생시켜야 할 경우도 생긴다.
# 파이썬은 raise라는 명령어를 이용해 오류를 강제로 발생시킬 수 있다.
# raise 되면 실행되던 함수는 곧바로 종료된다.

#  Bird라는 클래스를 상속받는 자식 클래스는
# 반드시 fly라는 함수를 구현하도록 만들고 싶은 경우(강제로 그렇게 하고 싶은 경우)가 있을 수 있다. 다음 예를 보자
# '구현(implement)' : 부모클래스의 메소드를 오버라이딩해서 재정의하는것.
```

```python
class Bird:
  def fly(self):
    print('fly() 호출')
    raise NotImplementedError
    print('fly() 종료')
```

```python
class Eagle(Bird):
  pass
```

```python
eagle = Eagle()

eagle.fly()
```

```python
# Bird 구현하기

class Eagle(Bird):
  # 오버라이딩
  def fly(self):
    print("fly() 독수리는 빠르게 납니다")
```

```python
eagle = Eagle()
eagle.fly()
```
