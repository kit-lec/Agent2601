# 다차원 리스트
- 리스트의 원소가 리스트
- 인덱스 여러개 사

```python
# 2차원 리스트.  -> 원소가 1차원 리스트
data = [
    [1, 2, 3],
    [10, 20, 30],
    [40, 50, 60],
]

data
```

```python
len(data)
```

```python
data[0]
```

```python
data[1]
```

```python
data[1][0]   # [][] 인덱스를 2개 사용하여 원소 사용.  2차원 리스트
```

```python
# data --> 2차원 list
# data[0] --> 1차원 list
# data[0][0] --> 0차원. scalar 값.

# ※ 인덱싱을 한다는 것은 차원 -1 축소 발생

# 2차원 리스트의 원소는 1차원 리스트
# 3..           원소는 2차원..
# 99차원...     원소는 98...
# n 차원 리스트의 원소는 n-1 차원 리스트
```

```python
data = [
    [1, 2, 3, 4, 5],
    [10, 20, 30, 40, 50],
    [11, 22, 33, 44, 55]
]

data
```

```python
# data 는 3 x 5  2차원 배열(행렬) 이라고도 함.
#       (3행 5열)
```

# Assignment Unpacking
tuple 은 iterable 객체다

대입연산자와 iterable 을 사용하여 여러 개의 변수에 대입 가능  (이를 Assignment unpacking 이라 함)

```python
# 사각형의 가로, 세로값
rec = [100, 200]
width = rec[0]
height = rec[1]

print(width, height)
```

```python
width, height = rec  # 변수들의 tuple 에 iterable 을 대입.

print(width, height)
```

# 🟦dict
- 딕셔너리
- key, value 쌍으로 구성.
- key 저장순서 유지됨.

```python
# dict : 딕셔너리

# dictionary 데이터 타입은  key-value 쌍으로 저장되는 데이터 집합이다.

# {Key1:Value1, Key2:Value2, Key3:Value3 ...}
# "이름" = "홍길동", "생일" = "몇 월 몇 일"   과 같은 자료형 담음

# 기존의 list, tuple 등은 value 중심

# Python 3.7 부터는 dict 의 key 저장순서로 유지됨.  순회할때에도 저장된 순서대로 리턴됨.
# key 는 중복 안된다.
```

```python
!python -V
```

```python
student = {"name":"손오공", "email":"son@mail.com"}

type(student)
student
```

```python
len(student) # key-value 쌍의 개수.
```

```python
student["name"]
```

```python
student.get("name")
```

```python
# student['age']  # KeyError
```

```python
student.get('age')  # key 가 없으면 None
```

```python
# get()  을 사용하면 예외적인 상황에서도
# 동작 가능하게 처리 가능
# student.get(key, default) : key 값이 없으면 default 값으로 리턴

student.get('age', 20)
```

```python
student
```

```python
student['name'] = '홍길동'  # 기존 key 값에 대입 (수정)

student
```

```python
student['address'] = '서울시' # 기존에 없던 key 값에 대입 (새로운 k-v 추가)

student
```

```python
# 삭제
del(student['email'])

student
```

```python
my_dict = {
    1: 'haha',
    2: 'hoho',
    2: 'HeHe',  # key 는 중복 불가!
    # [10, 20]: 3.14,   # list, set, dict 등은 dict 의 key 로 사용 불가!
    "two": {
        3.14: "pi",
        "pi": 3.14,
    },
    False: [10, 20, 30],
    (1, 2): 'nice',
}

my_dict
```
