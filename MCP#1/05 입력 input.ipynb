# 연습: 단위변환

```python
# 1야드(yd)는 91.44cm이고 1인치(in)는 2.54cm이다.
# 처음에는 야드를 입력받고, 두번째는 인치를 입력받아
# 각각 cm로 변환하여 다음 형식에 맞추어 소수 둘째자리까지 출력하시오.​

# [실행예]
# yard 입력: 23.45
# inch 입력: 41.273
# 23.45 yard 는 2144.27cm
# 41.27 inch 는 104.83cm
```

```python
# by. 조남호
ONE_YARD_TO_CM = 91.44
ONE_INCH_TO_CM = 2.54

yard_input = float(input('yard 입력: '))
inch_input = float(input('inch 입력: '))

print(f"{yard_input:.2f} yard 는 {yard_input * ONE_YARD_TO_CM:.2f}cm")
print(f"{inch_input:.2f} inch 는 {inch_input * ONE_INCH_TO_CM:.2f}cm")
```

```python
# by. 이은송
yard = float(input("야드를 입력하세요: "))
inch = float(input("인치를 입력하세요: "))

print(f"{yard} yard는 {yard*91.44:.2f}cm")
print(f"{inch} inch는 {inch*2.54:.2f}cm")
```
