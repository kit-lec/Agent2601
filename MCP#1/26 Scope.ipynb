# Scope

```python
# scope: 변수명, 함수명, 클래스명 등... 의 이름이 사용될수 있는 영역.
#   위 이름(name)을 어디에서 사용할 수 있느냐 에 대한 규칙.
#   변수가 살아있는 범위
```

# global variable, global scope
함수 바깥에서 선언한 변수.  

스크립트 전체에서 접근할 수 있는 변수를 전역 변수(global variable)라고 부릅니다.

특히 전역 변수에 접근할 수 있는 범위를 전역 범위(global scope)라고 합니다

```python
x = 10  # 전역 변수

def foo():
  print(x)  # 함수 안에서도 전역변수 '사용' 가능. 읽기 동작 ★

foo()
print(x)
```

# local variable, local scope

```python
# 지역 변수(local variable)  local scope
# 지역 변수는 변수를 만든 함수 안에서만 접근할 수 있고,
# 함수 바깥에서는 접근할 수 없습니다. 지역 변수가 접근할 수 있는 범위를 지역 범위(local scope)라 한다
```

```python
def foo():
  y = 10  # y 는 foo 의 지역변수
  print(y)

foo()  # 함수 종료(리턴) 과 함께 지역변수는 소멸.
print(y)
```

```python
# 함수 안에서 전역변수를 수정(쓰기 동작) 하려 한다면?

z = 10

def foo():
  z = 20  # z 는 foo의 지역변수다!!  쓰기동작★.
  print(z) # 지역변수 z 를 출력

foo() # 20
print(z) # 10
```

```python
a = 1

def foo():
  a = a + 1  # a는 (쓰기동작) foo 의 지역변수로 선언되지만...  a + 1 할 시점에서 a 값은 대입되있지 않은 상태
  print('foo] a=', a)

foo()
```

###  함수를 통해 전역변수 값을 '변경'하고 싶다면!
1. 방법1 : return 값 활용 (추천)
1. 방법2 : **global** 키워드 사용 (비추)

```python
a = 1  # 전역변수

def foo(a):  # 매개변수 a (지역변수)
  a = a + 1
  return a

a = foo(a)   # 리턴값으로 전역변수 a 를 변경한다.
print('전역 a=', a)
```

```python
a = 1

def foo():
  global a  # 이 함수 안에서 사용하는 a 는 전역변수 a 임을 선언.
  a = a + 1
  print('foo] a=', a)

foo()
print('전역 a=', a)
```

# 네임스페이스 (name space)
'이름'들이 사용되는 공간

```python
locals()  # 전역에서 호출하면 전역 이름공간
```

```python
locals().keys()
```

```python
def foo():
  cc, bb, dd = 10, 20, 40
  print(locals())   # foo() 함수 의 로컬 네임스페이스

foo()
```

```python
# 파이썬의 스코프 규칙: LEGB

# 이름을 찾을때 L → E → G → B 순서로 찾는다

# L (Local)     - 함수 안에서 정의된 변수
# E (Enclosing) - 중첩 함수에서 바깥 함수의 변수
# G (Global)    - 전역 레벨 변수
# B (Built-in)  - 내장된 이름
```

```python
k = 100
def aaa():
  k = 200
  print(k)

  def bbb():
    print(k)

  bbb()

aaa()
print(k)
```

# if, for, while, try, with 블럭(등)에서의 scope

```python
# ★★Python에서는 if, for, while, try, with 블럭(등)에서 새로운 scope를 만들지 않습니다.★★

# 변수는 가장 가까운 함수 단위의 지역 scope를 따르며, 함수 밖에서는 전역(global) scope를 가집니다.
# 즉! 파이썬은 for while if 블럭 안에 선언한 것들을 블럭 밖에서 사용 가능!

# 이 특성은 Java, C, JavaScript의 블럭 스코프(block scope)와는 다른 점입니다.
# 필요하면, nonlocal, global 키워드를 사용해 scope를 조절할 수도 있습니다.

# ▶파이썬의 scope 정리
# for, while, if 등 제어문 블럭	=> 함수 전체 또는 전역 scope
# def, lambda, class  =>	새로운 scope 생성
```

```python
for i in range(3):
  xx = i * 10  # xx 는 전역변수

xx  # for 바깥에서도 사용 가능.
```

```python
if True:
  yy = 100

yy
```

```python
try:
  ff = 100
  print('try: ff=', ff)
except:
  print('except: ff=', ff)
finally:
  print('finally: ff=', ff)

print('global ff=', ff)
```

```python
k = 100
def aaa():
  k = 200
  print(k)

  def bbb():
    nonlocal k  # k 는 bbb() 의 로컬변수가 아니라,  바깥쪽 감싸는 함수의 k 를 사용
    k = k + 1
    print(k)

  bbb()

aaa()
print(k)
```
