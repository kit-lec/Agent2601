```python
import os
key = os.getenv('SEOUL_API_KEY')
```

# 요청할 URL 준비

```python
date = '20260131'
end_index = 10
```

```python
url = f'http://openapi.seoul.go.kr:8088/{key}/json/CardSubwayStatsNew/1/{end_index}/{date}'
```

# requests 모듈
- http(웹 통신) 요청 - 응답

```python
import requests
```

```python
# get(url) : url 로 서버에 GET 방식으로 요청 -> Response 객체 리턴

response = requests.get(url)

response
```

```python
response.text
```

```python
# JSON response -> 파이썬 데이터로 변환
data = response.json()

data
```

```python
rows = data['CardSubwayStatsNew']['row']

len(rows)
```
