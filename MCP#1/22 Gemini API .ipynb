## 기본 호출

```python
from google import genai

client = genai.Client()

# 모델 호출
response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="Google Gemini 에 대해 설명해줘",
)

print(response.text)
```

**Google Gemini(구글 제미나이)**는 구글(Google)이 개발한 최첨단 인공지능(AI) 모델이자, 이를 기반으로 한 챗봇 서비스의 브랜드 이름입니다.

기존의 챗봇이었던 '바드(Bard)'가 업그레이드되면서 '제미나이'라는 이름으로 통합되었습니다. 제미나이에 대해 핵심적인 내용을 정리해 드릴게요.

---

### 1. 주요 특징: "네이티브 멀티모달(Native Multimodal)"
제미나이의장장 큰 특징은 처음부터 **멀티모달**로 설계되었다는 점입니다.
*   **멀티모달이란?** 텍스트뿐만 아니라 이미지, 영상, 오디오, 코드 등 다양한 형태의 정보를 동시에 이해하고 처리할 수 있는 능력을 말합니다.
*   기존 AI는 텍스트 모델에 이미지 인식 기능을 나중에 합친 방식이 많았지만, 제미나이는 설계 단계부터 모든 데이터를 동시에 학습하여 훨씬 자연스럽고 정확하게 정보를 처리합니다.

### 2. 모델 종류 (사이즈별 구분)
구글은 사용 목적에 따라 제미나이를 여러 버전으로 나누어 제공합니다.
*   **Gemini Ultra (울트라):** 가장 크고 복잡한 작업(코딩, 논리적 추론 등)을 수행하는 최고 성능 모델입니다. (유료 서비스인 '제미나이 어드밴스드'에 적용)
*   **Gemini Pro (프로):** 광범위한 작업에 최적화된 모델로, 속도와 성능의 밸런스가 좋습니다. 현재 무료 버전 제미나이에서 주로 사용됩니다.
*   **Gemini Flash (플래시):** 속도와 효율성에 최적화된 가벼운 모델로, 빠른 응답이 필요한 서비스에 적합합니다.
*   **Gemini Nano (나노):** 스마트폰(온디바이스 AI)에서 인터넷 연결 없이도 돌아갈 수 있도록 만든 가장 가벼운 모델입니다. (예: 갤럭시 S24, 픽셀 8 등 탑재)

### 3. 주요 강점
*   **방대한 컨텍스트 윈도우:** 특히 'Gemini 1.5 Pro' 같은 모델은 수십만 줄의 코드, 몇 시간 분량의 영상, 수천 페이지의 문서를 한꺼번에 읽고 분석할 수 있는 능력이 매우 뛰어납니다.
*   **구글 생태계와의 연결:** 구글 문서(Docs), 지메일(Gmail), 구글 드라이브, 유튜브 등과 연동하여 내 정보를 찾거나 요약해달라고 요청할 수 있습니다.
*   **논리 및 코딩:** 수학 문제 풀이, 복잡한 프로그래밍 코드 작성 및 디버깅 능력이 매우 강력합니다.

### 4. 사용 방법
*   **웹:** [gemini.google.com](https://gemini.google.com)에 접속하여 대화할 수 있습니다.
*   **모바일:** 안드로이드에서는 'Gemini' 앱을 다운로드하거나 기존 구글 어시스턴트를 대체하여 사용할 수 있고, iOS(아이폰)에서는 구글 앱 내의 제미나이 탭을 통해 사용할 수 있습니다.
*   **업무 도구:** 구글 워크스페이스(Google Workspace)를 사용하면 문서 작성이나 이메일 초안 작성 시 제미나이의 도움을 받을 수 있습니다.

### 요약

## streaming
- 전체 결과물이 완성되기까지 시간이 걸린다
- streaming 을 사용하면 단어들(토큰들)이 생성될때마다 출력시킬수 있다.

```python
from google import genai

client = genai.Client()

# 모델 호출
response = client.models.generate_content_stream(  # generate_content_stream 는 iterable 리턴
    model="gemini-3-flash-preview",
    contents="Google Gemini 에 대해 설명해줘",
)

for chunk in response:
  print(chunk.text, end="")
```

## System instruction
- gemini 모델의 답변 무드나 동작을 지시해줄수 있다.
- config= 에 GenerateContentConfig  객체로 이를 전달해준다.

```python
from google import genai
from google.genai import types

client = genai.Client()

# 모델 호출
response = client.models.generate_content(
    model="gemini-3-flash-preview",

    config= types.GenerateContentConfig(
        system_instruction="너는 고양이다.  너의 이름은 '냐코' 다"
    ),

    contents="안녕?",
)

print(response.text)
```

## temperature=
텍스트를 생성할 때 얼마나 창의적이고 무작위적인 답변을 낼지 조절하는 변수

```python
from google import genai
from google.genai import types

client = genai.Client()

# 모델 호출
response = client.models.generate_content(
    model="gemini-3-flash-preview",

    config= types.GenerateContentConfig(
        temperature=0.1
    ),

    contents="AI 가 어떻게 동작하는지 알려줘",
)

print(response.text)
```

## multi modal
- text 뿐 아니라, 이미지, 영상 등의 미디어와 함께 추론 가능.

```python
import requests
from PIL import Image
from io import BytesIO

url = "https://cdn.britannica.com/18/129118-050-BF48319E/organ.jpg"

response = requests.get(url)  # 이미지 데이터 요청

image = Image.open(BytesIO(response.content))

image
```

```python
from google import genai

client = genai.Client()

# 모델 호출
response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=[image, "이 악기에 대해 설명해 줘"],
)

print(response.text)
```
