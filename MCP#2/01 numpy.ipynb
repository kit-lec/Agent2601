# shape (차원) 변경하기
**차원변환**은 데이터분석 과 인공지능에서 매우 빈번하게 발생되는 작업이니만큼 자유자재로 변환할수 있어야 하고, 머리속으로 내가 변환하는 데이터의 구조가 그려지도록 익숙해져야 합니다

사용 예) 이미지 데이터 벡터화 - 이미지는 기본적으로 2차원 혹은 3차원(RGB)이나 트레이닝을 위해 1차원으로 변경하여 사용 됨

## ravel(), np.ravel()
  - [ˈrævl]
  - 다차원배열을 1차원으로 변경 (흔히 '**펼친다**'라고 말함)
  - 'order' 파라미터
    - 'C' - row 우선 변경 ,   C style
    - 'F - column 우선 변경,   Fortran style
  - ravel() 은 np 안에 일반 함수로도 있고, 혹은 ndarray 의 멤버함수로서도 존재

## flatten()
 - 다차원 배열을 1차원으로 변경
 - ravel과의 차이점:
     - copy를 생성하여 변경함(즉 원본 데이터가 아닌 복사본을 반환)
     - ndarray 의 멤버함수로만 제공됨
 - 'order' 파라미터
   - 'C' - row 우선 변경
   - 'F - column 우선 변경

```python
x
```

```python
x.ravel()  # (3, 5) -> (15,)
```

```python
np.ravel(x)
```

```python
x.flatten()
```

## reshape 함수
 - array의 shape을 다른 차원으로 변경
 - 주의할점은 reshape한 후의 결과의 전체 원소 개수(size)와 이전 개수가 같아야 가능
 - 사용 예) 이미지 데이터 벡터화 - 이미지는 기본적으로 2차원 혹은 3차원(RGB)이나 트레이닝을 위해 1차원으로 변경하여 사용 됨
 - **-1** 값을 주어 나머지 차원을 유추하게 할수 있다

```python
x2 = np.arange(36)
x2
```

```python
x2.reshape(6, 6)
```

```python
# -1 의 차원 크기가 유추되어 reshape 된다  => (6, 6)
x2.reshape(6, -1)
```

```python
x2.reshape(6, -1, 2)  # (6, 3, 2)
```

## 차원 확장 / 제거

- 차원 확장/제거 하는 동작도 머신러닝에서 많이 사용된다.
- **reshape()** 로 차원 변환(확장/제거) 자유롭게 가능

### 차원 확장 : np.expand_dims()
- axis= 로 지정된 '차원을 추가'한다.


### 차원 자동 제거 : squeeze, np.squeeze()

- 차원 중 사이즈가 1인 것을 찾아 스칼라값으로 바꿔 해당 '차원을 제거'한다.

```python
x3 = np.arange(6)  # (6,)
x3
```

```python
# (6,) => (6, 1) 로 차원 확장
x3 = x3.reshape(6, 1)
x3
```

```python
x3 = x3.reshape(6, 1, 1)
x3
```

```python
# 1차원으로 축소 (차원 제거)
x3.reshape((6,))
```

```python
# squeeze() 사용. 차원크기가 1인 차원을 제거  (6, 1, 1) => (6,)
x3.squeeze()
```

```python
x4 = np.arange(6).reshape((2, 1, 3, 1))
x4
```

```python
# (2, 1, 3, 1) => (2, 3)
x4.squeeze()
```

```python
x5 = np.arange(3)  # (3,)
x5
```

```python
# expand_dims 사용: 0번째 차원축(axis) 에 새로운 차원 추가(삽입)
y5= np.expand_dims(x5, axis=0)   # (3,) -> (1, 3)
print(y5.shape)
y5
```

```python
x6 = np.arange(4).reshape(2, 2)
x6
```

```python
# 마지막에 차원축 추가
# (2, 2) => (2, 2, 1)

y7 = np.expand_dims(x6, -1)
print(y7.shape)
y7
```

```python
# axis 0 과 axis -1 에 새로운 차원 추가
# (2, 2) -> (1, 2, 2, 1)
y7 = np.expand_dims(x6, (0, -1))
print(y7.shape)
y7
```

```python
x8 = np.array([2, 0, 1, 8])  # (4,)
x8
```

```python
# np.newaxis 사용
y8 = x8[np.newaxis, :]  # (4,) -> (1, 4)
print(y8.shape)
y8
```

```python
# ↑ 1차원 벡터를 row vector 로 변환한다 라고도 함.
```

```python
# np.newaxis 사용. (4,) => (4, 1)
a2 = x8[:, np.newaxis]
print(a2.shape)
a2
```

## axis (차원축)

```python
# numpy 의 많은 함수들은 axis=  파라미터를 갖고 있습니다.
# axis (축) 은 각 '차원' 을 의미합니다.
# 1차원 array 라면  각 차원에 대한 axis 값은 0  <- 1개 입니다.
# 2차원 array 라면  각 차원에 대한 axis 값은 0, 1  <- 2개 입니다.
# 3차원 array 라면  각 차원에 대한 axis 값은 0, 1, 2  <- 3개 입니다.

# axis= 값을 명시하면 그 함수의 연산 은 해당 axis(축) 에 '따라서' 연산을 수행함

# 과연 '따라서' 의 의미가 무엇인지 실습을 통해 이해해봅시다
```

```python
x1 = np.arange(15)
x1
```

```python
np.sum(x1)
```

```python
np.sum(x1, axis=None)
```

```python
x2 = x1.reshape(3, 5)
x2
```

```python
x2.shape

# (행, 열)
# (axis 0, axis 1)
```

```python
np.sum(x2) # axis=None
```

```python
np.sum(x2, axis=0)  # (3, 5) => (5,)

# 15 = 0 + 5 + 10
#     [0,0] + [1,0] + [2,0]
```

```python
x2
```

```python
np.sum(x2, axis=1) # (3, 5) => (3,)

# 35 = 5  +  6  + 7  + 8  + 9
#    [1,0] + [1,1] + [1,2] + [1,3] + [1,4]
```

```python
np.sum(x2, axis=-1)
```

## 도전!  서울시 5월 일자별 평균기온

```python
temp = np.array(
        [23.9, 24.4, 24.1, 25.4, 27.6, 29.7,
         26.7, 25.1, 25.0, 22.7, 21.9, 23.6,
         24.9, 25.9, 23.8, 24.7, 25.6, 26.9,
         28.6, 28.0, 25.1, 26.7, 28.1, 26.5,
         26.3, 25.9, 28.4, 26.1, 27.5, 28.1, 25.8])

"""

도전]
Q1] 평균기온이 25도를 넘는 날의 수는?  => 21일
Q2] 평균기온이 25도를 넘는 날의 평균기온은? => 26.8도

.mean()
.sum()

"""
None
```

```python
# by 조남호

daysAbove25dc = temp[temp > 25].size

avgTempAbove25dc = temp[temp > 25].mean() # 26.857... => 26.9
# 26.857 * 10 => 268.57 => trunc 268 / 10 => 26.8

print(f"평균기온이 25도를 넘는 날의 수는?  => {daysAbove25dc}일")
print(f"평균기온이 25도를 넘는 날의 평균기온은? => {int(avgTempAbove25dc * 10) / 10}도")
```

```python
# By 우준영
print("Q1] 평균기온이 25도를 넘는 날의 수는?")
result1=temp[temp>25]
print(len(result1))

print("Q2] 평균기온이 25도를 넘는 날의 평균기온은? => 26.8도")
print(f"{result1.sum() / len(result1):.1f}")
```
