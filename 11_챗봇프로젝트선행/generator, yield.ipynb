# generator

```python
def number_generator():
    yield 0  # 함수 안에서 yield 기술하면 이 함수는 generator 가 됨.
    yield 1
    yield 2
```

```python
number_generator()  # generator object
```

```python
# generator 는 iterator 를 생성해주는 함수
```

```python
for i in number_generator():
    print(i)
```

```python
def number_generator():
    print('🐹 gen1')
    yield 0
    print('🐹 gen2')
    yield 1
    print('🐹 gen3')
    yield 2

for i in number_generator():
    print('📌for', i)
```

왜 **yield** 라고 이름 지었을까?  양보하다란 뜻<br>
즉 **yield 값** 으로 함수 바깥으로 값을 전달하면서 코드 실행을 함수 바깥에 양보,<br>
현재 함수는 잠시 중단하고 함수 바깥의 코드가 실행됨.
